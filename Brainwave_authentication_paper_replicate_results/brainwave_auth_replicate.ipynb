```python
dir_path = 'Dataset'
raw_dict = {}

for filename in os.listdir(dir_path):
    if filename.endswith('.edf'):
        file_path = os.path.join(dir_path, filename)
        raw = mne.io.read_raw_edf(file_path, preload = True)
        raw.filter(1, 50, fir_design = 'firwin')
        subject_number = filename[6:9]
        new_key = f"subject_{int(subject_number):02d}"
        raw_dict[new_key] = raw
```

```python
def extract_number(filename):
    match = re.search(r'\d+', filename)
    if match:
        return int(match.group(0))
    return 0

sorted_items = sorted(raw_dict.items(), key = lambda item: extract_number(item[0]))
raw_dict = OrderedDict(sorted_items)

print(f'sorted dictionary keys: {raw_dict.keys()}')
print(f'raw_dict size: {len(raw_dict)}')
```

```python
event_times_dict = {}

sfreq = 256

for key in event_dict:
    i = extract_number(key)
    if i is None:
        continue

    # Create an empty list to store the events
    events = []

    #
    for row in event_dict[key]:
        # Converts latency from seconds to samples
        sample_index = int(row['Latency'] * sfreq)
        
        # Define the Event Code
        if 'S1P300' in row['Type']:
            event_code = 1
        elif 'S2P300' in row['Type']:
            event_code = 2
        elif 'N400words' in row['Type']:
            event_code = 3
        elif 'N400sent' in row['Type']:
            event_code = 4
        elif 'Faces' in row['Type']:
            event_code = 5
        else:
            # If the event is not one of the above, skip it
            continue
        
        # Append the event to the list
        events.append([sample_index, 0, event_code])

    # Convert the list to a numpy array
    exec(f'events_{i} = np.array(events, dtype = int)')
    
    # Store the array in a dictionary
    exec(f'event_times_dict["{key}"] = events_{i}')
```

```python
def remove_insufficient_epochs(power_spectrum_features_dict):
    subjects_to_delete = []
    
    for subject in subject_event_dict:
        for event_id, epochs in subject_event_dict[subject].items():
            if epochs.shape[0] < 3:
                print(f'Deleting {subject} event: {event_id} due to insufficient epochs: {epochs.shape[0]}')
                subjects_to_delete.append((subject, event_id))
                
    for subject, event_id in subjects_to_delete:
        del subject_event_dict[subject][event_id]
        
    return subject_event_dict

subject_event_dict = remove_insufficient_epochs(subject_event_dict)
```

```python
# Calculating the AR coefficients
def calc_ar_coeffs(epochs, order):
    ar_coeffs_dict = {}
    
    for subject in subject_event_dict:
        ar_coeffs_dict[subject] = {}
        
        for event_id, epochs_data in subject_event_dict[subject].items():
            n_epochs, n_channels, n_times = epochs_data.shape
            ar_coeffs = np.empty((n_epochs, n_channels, order + 1))

            for i in range(n_epochs):
                for j in range(n_channels):
                    channel_data = epochs_data[i][j]
                    model = AutoReg(channel_data, lags = order)
                    results = model.fit()
                    ar_coeffs[i][j] = results.params

            ar_coeffs_dict[subject][event_id] = ar_coeffs
    
    return ar_coeffs_dict

order = 10
ar_coeffs_dict = calc_ar_coeffs(subject_event_dict, order)
```

```python
def calc_power_spectrum(epoch, fs):
    #print(f'epoch received shape: {epoch.shape}')
    
    n_channels = epoch.shape[0]
    
    #print(f'n_channels: {n_channels}')
    
    freq_bands = [(1, 10), (10, 13), (13, 30), (30, 50)]
    power_spectrum_features = np.empty((n_channels, len(freq_bands)))
    
    for i in range(n_channels):
        #print(f'\nepoch with all channels shape: {epoch.shape}')
        #print(f'epoch channel {i} shape: {epoch[i].shape}')
        #print(f'fs: {fs}')
        
        freqs, psd = welch(epoch[i][np.newaxis, :], fs, nperseg = epoch.shape[1], axis = 1)
        
        #print(f'freqs shape: {freqs.shape}')
        #print(f'psd shape: {psd.shape}')
        
        for j, (fmin, fmax) in enumerate(freq_bands):
            idx_band = np.logical_and(freqs >= fmin, freqs <= fmax)
            #print(f'freq range: {fmin} - {fmax} Hz')
            #print(f'idx_band: {idx_band}')
            power_spectrum_features[i, j] = np.sum(psd[:, idx_band])
            #print(f'power_spectrum_features [i, j]: {power_spectrum_features[i, j]}')
            
    return power_spectrum_features

power_spectrum_features_dict = {}

for subject in subject_event_dict:
    power_spectrum_features_dict[subject] = {}
    
    for event_id, epochs_data in subject_event_dict[subject].items():
        n_epochs = epochs_data.shape[0]
        #print(f'n_epochs: {n_epochs}\n\n\n')
        power_spectrum_features = np.empty((n_epochs, 14, 4))

        for i in range(n_epochs):
            epoch_data = epochs_data[i]
            #print(f'\n\nepochs being sent: {i} shape: {epoch_data.shape}\n\n')
            power_spectrum_features[i] = calc_power_spectrum(epoch_data, fs = sfreq)

        power_spectrum_features_dict[subject][event_id] = power_spectrum_features
    
```

```python
def scale_ar_coeffs(ar_coeffs_dict):
    ar_scaler_dict = {}
    scaled_ar_coeffs_dict = {}

    for subject in ar_coeffs_dict:
        scaled_ar_coeffs_dict[subject] = {}
        ar_scaler_dict[subject] = {}

        for event_id, ar_coeffs in ar_coeffs_dict[subject].items():
            # Reshape for scaling
            original_shape = ar_coeffs.shape
            reshaped_features = ar_coeffs.reshape(original_shape[0], -1)
            
            # Scale features
            ar_scaler = StandardScaler()
            scaled_ar_coeffs = ar_scaler.fit_transform(reshaped_features)

            scaled_ar_coeffs_dict[subject][event_id] = scaled_ar_coeffs
            ar_scaler_dict[subject][event_id] = ar_scaler

    return scaled_ar_coeffs_dict, ar_scaler_dict

scaled_ar_coeffs_dict, ar_scaler_dict = scale_ar_coeffs(ar_coeffs_dict)
```

```python
splits = {}

for subject in combined_features_dict:
    splits[subject] = {}
    
    for event_id in combined_features_dict[subject]:
        x = combined_features_dict[subject][event_id]
        y = [event_id] * len(x)

        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state = 42)
        
        #scaler = StandardScaler()
        
        #x_train = scaler.fit_transform(x_train)
        #x_test = scaler.transform(x_test)
        
        splits[subject][event_id] = (x_train, x_test, y_train, y_test)
```

```python
svm = SVC(kernel = 'linear', C = 1, random_state = 42)
eer_values = defaultdict(list)

cv = RepeatedStratifiedKFold(n_splits = 3, n_repeats = 10, random_state = 42)

for subject in splits:
    for event_id in splits[subject]:
        
        x_train_authenticate = splits[subject][event_id][0]
        y_train_authenticate = ['authenticate'] * len(x_train_authenticate)

        x_train_reject = []
        y_train_reject = []

        for other_subject in splits:
            if other_subject != subject:
                for other_event_id in splits[other_subject]:
                    if other_event_id == event_id:
                        x_train_reject.extend(splits[other_subject][other_event_id][0])
                        y_train_reject.extend(['reject'] * len(splits[other_subject][other_event_id][0]))

        x_train = np.concatenate((x_train_authenticate, x_train_reject), axis = 0)
        y_train = y_train_authenticate + y_train_reject
            
        for train_index, test_index in cv.split(x_train, y_train):
            x_train_cv, x_test_cv = x_train[train_index], x_train[test_index]
            y_train_cv, y_test_cv = np.array(y_train)[train_index], np.array(y_train)[test_index]

            if len(np.unique(y_train_cv)) < 2:
                continue
            
            scaler = StandardScaler()
            
            x_train_cv = scaler.fit_transform(x_train_cv)
            x_test_cv = scaler.transform(x_test_cv)
            
            svm.fit(x_train_cv, y_train_cv)
            y_score = svm.decision_function(x_test_cv)

            fpr, tpr, thresholds = roc_curve(y_test_cv, y_score, pos_label = 'authenticate')
            frr = 1 - tpr
            #print(f'fnr: {fnr}')
            #print(f'fpr: {fpr}')
            if 'authenticate' not in y_test_cv:
                continue
            eer_threshold = thresholds[np.nanargmin(np.absolute((frr - fpr)))]
            eer = fpr[np.nanargmin(np.absolute((frr - fpr)))]
            eer_values[event_id].append(eer)
        
    #eer_values.append(np.mean(eer_subject_values))
average_eer_values = {event_id: np.mean(eers) for event_id, eers in eer_values.items()}    

#print(f'eer values: {eer_values}')
#average_eer = np.nanmean(eer_values)
#print(f'Average EER per event: {average_eer_values}')
```

```python
# Define the classifiers
classifier_dict = {
    'SVM': SVC(kernel='linear', C=1, random_state=42),
    'LG': LogisticRegression(random_state=42),
    'RF': RandomForestClassifier(random_state=42),
    'KNN': KNeighborsClassifier(),
    'GNB': GaussianNB(),
    'LDA': LinearDiscriminantAnalysis()
}

# Initialize a dictionary to store EER values for each classifier
eer_values = {classifier: defaultdict(list) for classifier in classifier_dict}

cv = RepeatedStratifiedKFold(n_splits=3, n_repeats=10, random_state=42)

for subject in splits:
    for event_id in splits[subject]:
        x_train_authenticate = splits[subject][event_id][0]
        y_train_authenticate = ['authenticate'] * len(x_train_authenticate)

        x_train_reject = []
        y_train_reject = []

        for other_subject in splits:
            if other_subject != subject:
                for other_event_id in splits[other_subject]:
                    if other_event_id == event_id:
                        x_train_reject.extend(splits[other_subject][other_event_id][0])
                        y_train_reject.extend(['reject'] * len(splits[other_subject][other_event_id][0]))

        x_train = np.concatenate((x_train_authenticate, x_train_reject), axis=0)
        y_train = y_train_authenticate + y_train_reject

        for name, classifier in classifier_dict.items():
            for train_index, test_index in cv.split(x_train, y_train):
                x_train_cv, x_test_cv = x_train[train_index], x_train[test_index]
                y_train_cv, y_test_cv = np.array(y_train)[train_index], np.array(y_train)[test_index]

                if len(np.unique(y_train_cv)) < 2:
                    continue

                scaler = StandardScaler()
                x_train_cv = scaler.fit_transform(x_train_cv)
                x_test_cv = scaler.transform(x_test_cv)

                classifier.fit(x_train_cv, y_train_cv)
                y_score = classifier.decision_function(x_test_cv) if hasattr(classifier, "decision_function") else classifier.predict_proba(x_test_cv)[:, 1]

                fpr, tpr, thresholds = roc_curve(y_test_cv, y_score, pos_label='authenticate')
                frr = 1 - tpr

                if 'authenticate' not in y_test_cv:
                    continue

                eer_threshold = thresholds[np.nanargmin(np.absolute((frr - fpr)))]
                eer = fpr[np.nanargmin(np.absolute((frr - fpr)))]
                eer_values[name][event_id].append(eer)

# Calculate the average EER for each classifier and event
average_eer_values = {classifier: {event_id: np.mean(eers) for event_id, eers in events.items()} for classifier, events in eer_values.items()}
print(f'Average EER per event for each classifier: {average_eer_values}')
```
